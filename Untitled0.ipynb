#    Projet 4IIR1 : Reconnaissance du langage des signes avec un réseau de neurones convolutif (CNN)

L’objectif de ce projet est de construire un modèle de deep learning capable de reconnaître des lettres et chiffres du langage des signes à partir d’images organisées dans un dataset présent sur kaagle.
Pour cela, nous allons utiliser un réseau de neurones convolutif (CNN), un type de réseau très performant pour le traitement d’images.

#                    **Étapes principales du projet**

    Récupération du dataset

-Nous allons télécharger un dataset depuis Kaggle en utilisant la bibliothèque kagglehub.

-Ce dataset contient des images de mains représentant différentes lettres du langage des signes.

    Chargement et préparation des données

-Nous allons organiser les images en jeux d’entraînement et de validation.

-Nous utiliserons tf.keras.utils.image_dataset_from_directory pour créer des tf.data.Dataset à partir des dossiers d’images.

-Nous définirons la taille des images, la taille des batchs, etc.

    Prétraitement & augmentation des données

-Normalisation des pixels (Toutes les images doient avoir les mêmes dimensions pour être traitées de la même manière par le CNN).

-Augmentation de données (rotations, flips, zooms) pour rendre le modèle plus robuste et éviter l’overfitting.

    Construction du modèle CNN
  On va empiler plusieurs couches :

-Conv2D pour extraire des caractéristiques (bords, formes, etc.).

-MaxPooling2D pour réduire la dimension des cartes de caractéristiques.

-Dense (couches entièrement connectées) pour la décision finale.

-Dropout pour limiter l’overfitting.

    Compilation & entraînement du modèle

1/Définition de la fonction de perte, de l’optimiseur et des métriques.
2/Entraînement sur plusieurs époques et observation de l’accuracy et de la perte.

    Visualisation des performances

-Courbes de précision (accuracy) et de perte (loss) sur entraînement / validation.

-Test sur une image

-Sélection d’une image.

-Prédiction de la classe.

-Vérification de la cohérence avec le label attendu.

    Bibliothèques utilisées

-kagglehub : permet de télécharger facilement des datasets depuis Kaggle sans devoir gérer manuellement les fichiers.

-tensorflow / keras : framework de deep learning qui contient tout ce qu’il faut pour construire, entraîner et utiliser des réseaux de neurones.

-matplotlib : pour tracer des courbes (accuracy / loss) et visualiser les performances du modèle.

-pathlib (Path) : pour manipuler les chemins de fichiers de façon propre et portable.

```python
# =========================
# 1. INSTALLATION & IMPORTS
# =========================

# Installation de kagglehub (si ce n'est pas déjà installé)
# Le "!" indique à Colab d'exécuter une commande système (comme dans un terminal).
!pip install kagglehub --quiet

# Import des bibliothèques nécessaires
import kagglehub               # Pour télécharger le dataset depuis Kaggle
import os                      # Pour manipuler les chemins / dossiers au niveau système
from pathlib import Path       # Pour gérer les chemins de manière propre et orientée objet

# =========================
# 2. TÉLÉCHARGEMENT DU DATASET
# =========================

# kagglehub.dataset_download("nom_du_dataset") télécharge le dataset demandé
# et retourne le chemin local où il a été enregistré.
path = kagglehub.dataset_download("harshvardhan21/sign-language-detection-using-images")

print("Chemin vers les fichiers du dataset :", path)

# Lister récursivement le contenu du dossier pour comprendre la structure
# -R : liste aussi les sous-dossiers
!ls -R "$path"
```

# **Interprétatoin:**
Après analyse du résultat de la commande "**!ls -R "$path"**" on remarque que le dataset est constitué de différets dossiers comportat chacu des images du chiffre/nombre cconcerné. Les données étant déà organisées cela simplifie nôtre démarche et l'étape suivante est celle de normalisation des données.

```python
# =========================
# 3. IMPORTS POUR LE DEEP LEARNING
# =========================

import tensorflow as tf                            # Framework de deep learning
from tensorflow import keras                       # Interface haut niveau de TensorFlow
from tensorflow.keras import layers                # Pour créer les différentes couches du modèle
import matplotlib.pyplot as plt                    # Pour tracer les courbes d'entraînement

# =========================
# 4. PARAMÈTRES GÉNÉRAUX
# =========================

# Taille des images en entrée du réseau
# On redimensionnera toutes les images à cette taille.
IMG_SIZE = (128, 28)
# Taille du batch : nombre d'images envoyées au réseau à chaque itération
BATCH_SIZE = 32
# Graine aléatoire pour rendre les résultats reproductibles
SEED = 123
```

```python
# =========================
# 5. DÉFINITION DU DOSSIER D'IMAGES
# =========================

# Path(path) transforme le chemin (string) en objet Path.
# On ajoute ensuite le sous-dossier qui contient les images
# ⚠️ ADAPTE "Sign_language" EN FONCTION DE LA STRUCTURE RÉELLE
data_dir = Path(path) / "data"

print("Dossier utilisé pour les images :", data_dir)

# Lister le contenu du dossier choisi pour vérifier qu'il contient bien des sous-dossiers de classes
!ls "$data_dir"
```

```python
# =========================
# 6. CRÉATION DES DATASETS D'ENTRAÎNEMENT ET DE VALIDATION
# =========================

# image_dataset_from_directory va :
#  - lire toutes les images dans data_dir
#  - attribuer une étiquette (label) en fonction du nom du sous-dossier (1, 2, 3, ..., A, B, ...)
#  - créer un objet tf.data.Dataset prêt pour l'entraînement

train_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,              # dossier racine qui contient un sous-dossier par classe
    validation_split=0.2,  # 20% des images seront réservées pour la validation
    subset="training",     # on demande ici la partie "entraînement"
    seed=SEED,             # graine pour que la séparation train/val soit toujours la même
    image_size=IMG_SIZE,   # redimensionne toutes les images à la même taille (64x64)
    batch_size=BATCH_SIZE  # nombre d'images par batch
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,  # même split 20%
    subset="validation",   # ici on demande la partie "validation"
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

# Récupération des noms de classes (dossiers) trouvés,
# par exemple : ['1', '2', ..., '9', 'A', 'B', ..., 'Z']
class_names = train_ds.class_names
print("Classes trouvées :", class_names)

# Nombre total de classes (important pour la dernière couche du réseau)
num_classes = len(class_names)
print("Nombre de classes :", num_classes)
```

```python
# =========================
# 7. OPTIMISATION DU PIPELINE DE DONNÉES
# =========================

# AUTOTUNE permet à TensorFlow de choisir automatiquement la meilleure stratégie
# pour précharger les données et optimiser l'utilisation du CPU/GPU.
AUTOTUNE = tf.data.AUTOTUNE

# Pour le jeu d'entraînement :
#  - shuffle(1000) mélange les exemples pour casser l'ordre des données.
#  - prefetch(AUTOTUNE) prépare à l'avance le batch suivant pendant que le modèle
#    est en train d'apprendre sur le batch courant (gain de temps).
train_ds = train_ds.shuffle(1000).prefetch(buffer_size=AUTOTUNE)

# Pour le jeu de validation :
#  - pas nécessaire de shuffle, on garde l'ordre
#  - on garde juste le prefetch pour accélérer un peu
val_ds = val_ds.prefetch(buffer_size=AUTOTUNE)
```

```python
# =========================
# 8. AUGMENTATION DE DONNÉES
# =========================

# L'objectif est de générer des "nouvelles" images à partir des images existantes
# en appliquant des transformations aléatoires. Cela :
#  - augmente artificiellement la taille du dataset
#  - rend le modèle plus robuste (moins d'overfitting)
# Cela permet au modèle d'avoir plusieur versions et d'éviter un apprentissage "par coeur"

from tensorflow import keras
from tensorflow.keras import layers

data_augmentation = keras.Sequential(
    [
        # Retourner l'image horizontalement (comme si la main était de l'autre côté)
        layers.RandomFlip("horizontal"),

        # Faire une petite rotation (positif ou négatif) jusqu'à 10% de tour
        layers.RandomRotation(0.1),

        # Zoomer légèrement à l'intérieur de l'image
        layers.RandomZoom(0.1),
    ]
)
```

```python
# =========================
# 9. CONSTRUCTION DU MODÈLE CNN
# =========================

# Définition de la forme d'entrée :
# (*IMG_SIZE, 3) = (64, 64, 3) -> hauteur, largeur, canaux (RGB)
inputs = keras.Input(shape=(*IMG_SIZE, 3))

# 1) Appliquer l'augmentation de données aux entrées
x = data_augmentation(inputs)

# 2) Normalisation des pixels : passage de [0, 255] à [0, 1]
x = layers.Rescaling(1./255)(x)

# ---- BLOC 1 : Convolution + MaxPooling ----
# Conv2D :
#  - 32 filtres (ou "kernels")
#  - taille du filtre 3x3
#  - activation ReLU pour introduire de la non-linéarité
#  - padding="same" pour garder la même taille (H, W)
x = layers.Conv2D(32, (3, 3), activation="relu", padding="same")(x)
# MaxPooling2D réduit la dimension spatiale de moitié (pooling 2x2)
x = layers.MaxPooling2D((2, 2))(x)

# ---- BLOC 2 : Convolution + MaxPooling ----
x = layers.Conv2D(64, (3, 3), activation="relu", padding="same")(x)
x = layers.MaxPooling2D((2, 2))(x)

# ---- BLOC 3 : Convolution + MaxPooling ----
x = layers.Conv2D(128, (3, 3), activation="relu", padding="same")(x)
x = layers.MaxPooling2D((2, 2))(x)

# Aplatir les cartes de caractéristiques en un vecteur 1D
x = layers.Flatten()(x)

# Dropout :
#  - pendant l'entraînement, 50% des neurones sont "éteints" aléatoirement
#  - ça force le réseau à ne pas trop dépendre d'un petit groupe de neurones
#  - réduit l'overfitting
x = layers.Dropout(0.5)(x)

# Couche entièrement connectée de 256 neurones
x = layers.Dense(256, activation="relu")(x)

# Couche de sortie :
#  - num_classes neurones (un par classe possible)
#  - activation softmax pour obtenir des probabilités sommant à 1
outputs = layers.Dense(num_classes, activation="softmax")(x)

# Création du modèle complet en reliant inputs -> outputs
model = keras.Model(inputs, outputs)

# Afficher un résumé du modèle
model.summary()
```

```python
# =========================
# 10. COMPILATION DU MODÈLE
# =========================

# optimizer="adam" :
#  - algorithme très utilisé et généralement efficace
# loss="sparse_categorical_crossentropy" :
#  - adaptée à la classification multi-classes
#  - "sparse" car les labels sont des entiers (0, 1, 2, ...) et non des vecteurs one-hot
# metrics=["accuracy"] :
#  - on surveille la précision pendant l'entraînement
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

# Nombre d'époques (passages complets sur le jeu d'entraînement)
EPOCHS = 15  # tu peux tester 20, 25, etc. si tu veux

# =========================
# 11. ENTRAÎNEMENT DU MODÈLE
# =========================

# model.fit entraîne le modèle :
#  - sur train_ds
#  - en évaluant à chaque époque sur val_ds
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS
)
```

```python
# =========================
# 12. VISUALISATION DES COURBES D'ENTRAÎNEMENT
# =========================

import matplotlib.pyplot as plt

# Récupérer l'historique de l'entraînement depuis l'objet "history"
acc = history.history["accuracy"]          # précision sur entraînement
val_acc = history.history["val_accuracy"]  # précision sur validation
loss = history.history["loss"]             # perte sur entraînement
val_loss = history.history["val_loss"]     # perte sur validation

epochs_range = range(EPOCHS)               # simplement [0, 1, ..., EPOCHS-1]

plt.figure(figsize=(12, 4))

# --- Courbe de précision ---
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label="Train Acc")
plt.plot(epochs_range, val_acc, label="Val Acc")
plt.legend()
plt.title("Accuracy (Précision)")
plt.xlabel("Époques")
plt.ylabel("Accuracy")

# --- Courbe de perte ---
plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label="Train Loss")
plt.plot(epochs_range, val_loss, label="Val Loss")
plt.legend()
plt.title("Loss (Perte)")
plt.xlabel("Époques")
plt.ylabel("Loss")

plt.show()
```

```python
# =========================
# 13. SAUVEGARDE DU MODÈLE
# =========================

# Cette commande enregistre :
#  - l'architecture du modèle
#  - les poids entraînés
#  - la configuration de compilation
model.save("sign_language_cnn.h5")

print("Modèle sauvegardé sous le nom : sign_language_cnn.h5")
```

```python
# =========================
# 14. TEST DU MODÈLE SUR UNE IMAGE
# =========================

import numpy as np
from tensorflow.keras.preprocessing import image

# On cherche une image au hasard dans le dossier data_dir.
# Ici on prend simplement la première image .jpg trouvée.
# ⚠️ Si les images ne sont pas en .jpg mais en .png, remplace "*.jpg" par "*.png"
img_path = str(list(data_dir.rglob("*.jpg"))[0])
print("Image de test :", img_path)

# Chargement de l'image et redimensionnement à IMG_SIZE (64x64)
img = image.load_img(img_path, target_size=IMG_SIZE)

# Conversion de l'image en tableau numpy
img_array = image.img_to_array(img)

# Ajout d'une dimension "batch" : on passe de (64, 64, 3) à (1, 64, 64, 3)
img_array = tf.expand_dims(img_array, 0)

# Normalisation comme pendant l'entraînement
img_array = img_array / 255.0

# Prédiction : on obtient un vecteur de probabilités pour chaque classe
predictions = model.predict(img_array)

# Classe prédite = indice de la probabilité maximale
predicted_index = np.argmax(predictions[0])

# Nom de la classe correspondante
predicted_class = class_names[predicted_index]

print("Classe prédite :", predicted_class)
print("Probabilités par classe :", predictions[0])
```
